```python
import numpy as np
import aspect_based_sentiment_analysis as absa
from IPython.core.display import display
```

```python
nlp = absa.load('absa/classifier-rest-0.1')
text = ("We are great fans of Slack, but we wish the subscriptions " 
        "were more accessible to small startups.")
document = nlp(text, aspects=["slack"])

(aspect_span,), = document
scores = np.round(aspect_span.scores, decimals=3)
```

```python
print(f'Sentiment for "slack": {repr(aspect_span.sentiment)}')
print(f'Scores (neutral/negative/positive): {scores}')
```

```python
html = absa.probing.explain(aspect_span)
display(html)
```
